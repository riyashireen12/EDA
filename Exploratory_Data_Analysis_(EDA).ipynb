```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
# 1. Summary Statistics
print("Summary Statistics:")
print(df.describe(include='all'))
```

```python
# 2. Null Values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# 3. Histograms for numeric features
df.hist(figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Columns")
plt.tight_layout()
plt.show()
```

```python
# 4. Boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Boxplot of Age and Fare")
plt.show()
```

```python
# 5. Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# 6. Pairplot
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
plt.show()
```

```python
# 7. Countplot for Categorical Data
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival Count by Sex")
plt.show()
```

```python
# 8. Detecting skewness
print("\nSkewness:")
print(df.skew(numeric_only=True))
```

```python
# 9. Multicollinearity check (using correlation)
print("\nPotential multicollinearity (correlation > 0.8):")
corr_matrix = df.corr(numeric_only=True)
print(corr_matrix[corr_matrix > 0.8])
```

```python
# 10. Interactive plot (optional using Plotly)
fig = px.scatter(df, x='Age', y='Fare', color='Survived', title='Age vs Fare by Survival')
fig.show()
```
